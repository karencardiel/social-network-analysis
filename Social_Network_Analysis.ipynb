##**Matriz de Adyacencia**

```python
# Obtener lista única de personas
people = sorted(set(user for conn in connections for user in conn))

# Crear mapeo de usuario a índice
user_to_index = {user: idx for idx, user in enumerate(people)}

# Inicializar la matriz de adyacencia
n = len(people)
adjacency_matrix = [[0] * n for _ in range(n)]

# Llenar la matriz con las conexiones
for user1, user2 in connections:
    idx1 = user_to_index[user1]
    idx2 = user_to_index[user2]
    adjacency_matrix[idx1][idx2] = 1
    adjacency_matrix[idx2][idx1] = 1  # Grafo no dirigido

# Imprimir la matriz de adyacencia con nombres alineados
col_width = max(len(name) for name in people) + 2
print("".ljust(col_width) + " ".join(name.ljust(col_width) for name in people))
for user, row in zip(people, adjacency_matrix):
    print(user.ljust(col_width) + " ".join(str(value).ljust(col_width) for value in row))
```

##**Algoritmo Floyd Warshall**

```python
def floyd_warshall(adjacency_matrix):
    n = len(adjacency_matrix)
    dist = adjacency_matrix.copy()

    # Inicializamos la distancia
    for i in range(n):
        for j in range(n):
            if dist[i][j] == 0 and i != j:
                dist[i][j] = np.inf  # Sin conexión entre nodos

    # Algoritmo de Floyd-Warshall
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][j] > dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist
```

```python
# Obtener la matriz de distancias
dist_matrix = floyd_warshall(adjacency_matrix)
```

##**Excentricidad**

```python
# Excentricidad de un Nodo: La excentricidad de un nodo es la distancia máxima desde ese nodo hasta cualquier otro nodo.
eccentricity = np.max(dist_matrix, axis=1)
```

```python
# Mostrar la excentricidad junto con los nombres de los usuarios
eccentricity_dict = dict(zip(people, eccentricity))
print("Excentricidad de los nodos:")
for user, ecc in eccentricity_dict.items():
    print(f"{user}: {ecc}")
```

```python
# Crear el grafo de excentricidad
G = nx.Graph()
for user1, user2 in connections:
    G.add_edge(user1, user2)

dist_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        if i != j:
            try:
                dist_matrix[i][j] = nx.shortest_path_length(G, people[i], people[j])
            except nx.NetworkXNoPath:
                dist_matrix[i][j] = np.inf

# Calcular la excentricidad
eccentricity = np.max(dist_matrix, axis=1)

# Añadir nodos y sus excentricidades
for user, ecc in zip(people, eccentricity):
    G.add_node(user, eccentricity=ecc)

# Obtener las excentricidades de los nodos
ecc_values = nx.get_node_attributes(G, 'eccentricity')

# Extraer el valor máximo de excentricidad para usarlo en la escala de colores
max_ecc = max(ecc_values.values())
min_ecc = min(ecc_values.values())
```

```python
node_color = []
for node in G.nodes():
    if node in center:  # Asignar color amarillo dorado para el centro
        node_color.append('#F39C12')  # Centro
    elif node in periphery:  # Asignar color púrpura para la periferia
        node_color.append('#A57BC4')  # Periferia
    elif node in [n for n, _ in top_closeness[:3]]:  # Top 3 en centralidad de cercanía
        node_color.append('#1ABC9C')  # Nodos intermedios
    else:
        node_color.append('#808080')  # Nodos en gris claro

node_size = [700 for _ in G.nodes()]  # Tamaño uniforme para todos los nodos

# Crear el grafo visualizando los nodos con colores personalizados
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color=node_color,  # Usando los colores personalizados
        node_size=node_size, font_size=10, font_weight='bold', edge_color='gray', alpha=0.8)

# Título
plt.title("ECCENTRICITY", fontsize=16)
plt.show()
```

##**Longitud Promedio del Camino**

```python
# Calcular la longitud promedio del camino
n = len(dist_matrix)  # Número de nodos
total_distance = 0
num_pairs = 0

# Recorrer la matriz de distancias
for i in range(n):
    for j in range(n):
        if i != j:  # No contar la distancia de un nodo consigo mismo
            total_distance += dist_matrix[i][j]
            num_pairs += 1

# Longitud promedio del camino
avg_path_length = total_distance / num_pairs
print(f'Longitud Promedio del Camino: {avg_path_length}')
```

##**Centro y Periferia de los Nodos**

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Suponiendo que ya tienes un grafo `G` y una lista `people` que contiene los nodos
# Vamos a calcular la excentricidad para cada nodo en el grafo
eccentricity = nx.eccentricity(G)

# Encontrar la excentricidad mínima y máxima
min_eccentricity = np.min(list(eccentricity.values()))
max_eccentricity = np.max(list(eccentricity.values()))

# Encontrar los nodos del centro (mínima excentricidad) y la periferia (máxima excentricidad)
center = [node for node, ecc in eccentricity.items() if ecc == min_eccentricity]
print(f'Centro del grafo: {center}')

periphery = [node for node, ecc in eccentricity.items() if ecc == max_eccentricity]
print(f'Periferia del grafo: {periphery}')

# Añadir nodos y sus excentricidades al grafo
for node, ecc in eccentricity.items():
    G.nodes[node]['eccentricity'] = ecc

# Crear lista de colores para los nodos
node_color = []
for node in G.nodes():
    if node in center:
        node_color.append('#F39C12')  # Centro en amarillo
    elif node in periphery:
        node_color.append('#A57BC4')  # Periferia en morado
    else:
        node_color.append('#808080')  # Otros nodos en gris

# Visualizar el grafo con nodos del centro y la periferia
plt.figure(figsize=(11, 8))
nx.draw(G, with_labels=True, node_color=node_color, node_size=700, font_size=10, font_weight='bold', edge_color='gray',
        alpha=0.7)
plt.title("GRAPH WITH CENTER AND PERIPHERY NODES")
plt.show()

# Calcular la centralidad de cercanía
closeness_centrality = nx.closeness_centrality(G)

# Mostrar la centralidad de cercanía ordenada
top_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)
print("Centralidad de cercanía de los nodos (ordenada):")
for user, centrality in top_closeness:
    print(f"{user}: {centrality}")
```
